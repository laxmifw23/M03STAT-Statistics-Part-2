# Theory Q&A

# 01. What is hypothesis testing in statistics
- Hypothesis testing is a statistical method used to determine whether a hypothesis about a population parameter is true or false. It involves formulating a null and alternative hypothesis, collecting sample data, and using statistical tests to decide whether to reject the null hypothesis based on the evidence.

# 02. What is the null hypothesis, and how does it differ from the alternative hypothesis
- The null hypothesis (H0) states that there is no significant difference or relationship, while the alternative hypothesis (H1) states that there is a significant difference or relationship. The null hypothesis is tested for possible rejection, whereas the alternative hypothesis is accepted if the null is rejected.

# 03. What is the significance level in hypothesis testing, and why is it important
- The significance level (a) is the probability of rejecting a true null hypothesis. It's a threshold for determining statistical significance, typically set at 0.05. Importance lies in balancing Type I errors (false positives) and ensuring reliable conclusions, as it determines when to reject the null hypothesis.

# 04. What does a P-value represent in hypothesis testing
- The P-value represents the probability of observing the test results (or more extreme) assuming the null hypothesis is true. A small P-value (typically < 0.05) indicates strong evidence against the null hypothesis, leading to its rejection, while a large P-value suggests insufficient evidence to reject the null.

# 05. How do you interpret the P-value in hypothesis testing
- A  P-value < 0.05 typically indicates strong evidence against the null hypothesis, suggesting a statistically significant result. A P-value > 0.05 indicates insufficient evidence to reject the null hypothesis, suggesting no statistically significant difference or relationship. The smaller the P-value, the stronger the evidence against the null.

# 06. What are Type 1 and Type 2 errors in hypothesis testing
- Type 1 error (a-error) occurs when a true null hypothesis is rejected, resulting in a false positive. Type 2 error (3-error) occurs when a false null hypothesis is not rejected, resulting in a false negative. Minimizing these errors is crucial for reliable conclusions, with a typically set to 0.05.

# 07. What is the difference between a one-tailed and a two-tailed test in hypothesis testing 
- A one-tailed test examines if a parameter is greater than or less than a value, testing directionality. A two-tailed test checks for any difference (greater or less than). One-tailed tests have more power to detect effects in one direction, while two-tailed tests are more conservative, testing both directions.

# 08.  What is the Z-test, and when is it used in hypothesis testing
- The Z-test is a statistical test used to determine if a sample mean differs significantly from a known population mean. It's used when the population standard deviation is known, sample size is large (n > 30), and data follows a normal distribution, typically for comparing means or proportions.

# 09. How do you calculate the Z-score, and what does it represent in hypothesis testing
- The Z-score is calculated as (sample mean - population mean) / (population standard deviation / âˆšsample size). It represents the number of standard deviations the sample mean is away from the population mean, indicating how many standard errors the sample mean is from the true mean.

# 10. What is the T-distribution, and when should it be used instead of the normal distribution
- The T-distribution is a probability distribution used for hypothesis testing when the sample size is small (n < 30) or the population standard deviation is unknown. It's similar to the normal distribution but has heavier tails, providing more conservative estimates and critical values for small samples or unknown variances.

# 11. What is the difference between a Z-test and a T-test
- The Z-test is used when the population standard deviation is known and sample size is large (n > 30). The T-test is used when the population standard deviation is unknown or sample size is small (n < 30). T-tests are more conservative and account for added uncertainty in small samples or unknown variances.

# 12.  What is the T-test, and how is it used in hypothesis testing
- The T-test is a statistical test used to compare the means of two groups or a sample mean to a known population mean. It's used for hypothesis testing when the population standard deviation is unknown or sample size is small (n < 30), determining if there's a significant difference between means.

# 13. What is the relationship between Z-test and T-test in hypothesis testing
- The Z-test and T-test are both used for hypothesis testing of means. The Z-test is used for large samples (n > 30) with known population standard deviation, while the T-test is used for small samples (n < 30) or unknown population standard deviation. Both tests compare sample means to population means.

# 14. What is a confidence interval, and how is it used to interpret statistical results
- A confidence interval is a range of values within which a population parameter is likely to lie. It's used to interpret statistical results by providing a margin of error and a level of confidence (e.g., 95%). The interval gives a sense of the estimate's precision and uncertainty.

# 15. What is the margin of error, and how does it affect the confidence interval
- The margin of error is the maximum amount by which a sample estimate may differ from the true population parameter. A smaller margin of error results in a narrower confidence interval, indicating more precise estimates. A larger margin of error results in a wider interval, indicating less precision.

# 16. How is Bayes' Theorem used in statistics, and what is its significance
- Bayes' Theorem updates the probability of a hypothesis based on new evidence. It's used in statistics to incorporate prior knowledge and observed data to make probabilistic inferences. Significance lies in its ability to revise probabilities as new information becomes available, enabling more accurate predictions and decision-making.

# 17. What is the Chi-square distribution, and when is it used
- The Chi-square distribution is a probability distribution used for hypothesis testing and confidence intervals for categorical data. It's used for tests of independence, goodness-of-fit, and homogeneity, typically when analyzing count data or comparing observed frequencies to expected frequencies under a null hypothesis, often with large samples.

# 18. What is the Chi-square goodness of fit test, and how is it applied
- The Chi-square goodness-of-fit test determines if observed categorical data fits a expected distribution. It's applied by comparing observed frequencies to expected frequencies under a null hypothesis, calculating the Chi-square statistic, and determining the p-value to assess significance, often used for testing distributional assumptions or theoretical expectations.

# 19. What is the F-distribution, and when is it used in hypothesis testing
- The F-distribution is a probability distribution used in hypothesis testing for comparing variances or testing regression models. It's commonly used in ANOVA (Analysis of Variance) to determine if variance between groups is significant, and in regression analysis to test overall model significance or compare nested models.

# 20. What is an ANOVA test, and what are its assumptions
- ANOVA (Analysis of Variance) is a statistical test comparing means across three or more groups. It assumes: normality of residuals, homogeneity of variance (equal variances), independence of observations, and random sampling. ANOVA determines if significant differences exist between group means, often used for comparing multiple treatment groups or conditions.

# 21. What are the different types of ANOVA tests
- There are several types of ANOVA tests:
1. One-way ANOVA: Compares means across one independent variable.
2. Two-way ANOVA: Examines the effect of two independent variables.
3. N-way ANOVA: Examines multiple independent variables.
4. Repeated Measures ANOVA: Compares means within subjects over time or conditions.
5. MANOVA (Multivariate ANOVA): Compares multiple dependent variables.
Each type serves specific research questions and designs.

# 22. What is the F-test, and how does it relate to hypothesis testing
- The F-test is a statistical test comparing variances or testing regression models. In hypothesis testing, it determines if variance between groups is significant or if a regression model is significant. The F-statistic calculates the ratio of explained variance to unexplained variance, assessing overall model fit or group differences.

```python
'''03. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution'''
import numpy as np

def bernoulli_pmf(x, p):
    """
    Calculate the probability mass function (PMF) of a Bernoulli distribution.

    Parameters:
    x (int): 0 or 1, the outcome of the Bernoulli trial
    p (float): probability of success (1)

    Returns:
    float: probability of the outcome
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        raise ValueError("x must be 0 or 1")

# Example usage:
p = 0.7  # probability of success
print("P(X=1) =", bernoulli_pmf(1, p))
print("P(X=0) =", bernoulli_pmf(0, p))
```

```python
'''09. Write a Python function to calculate Z-scores from a dataset and plot them'''
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Generate a dataset
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=100)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot the data and Z-scores
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, alpha=0.7, color='blue')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, alpha=0.7, color='green')
plt.title('Z-scores')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Identify outliers (Z-score > 2 or < -2)
outliers = data[np.abs(z_scores) > 2]
print("Outliers:", outliers)
```

```python
'''15. Implement hypothesis testing using Z-statistics for a sample dataset'''
import numpy as np
from scipy.stats import norm

# Parameters
sample_mean = 520
population_mean = 500
population_std = 100
sample_size = 100

# Calculate Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

# Significance level (Î±)
alpha = 0.05

print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
'''16. Create a confidence interval for a dataset using Python and interpret the result'''
import numpy as np
from scipy import stats

# Generate a dataset
np.random.seed(0)
data = np.random.normal(loc=5, scale=2, size=100)

# Calculate the sample mean and standard error
sample_mean = np.mean(data)
std_error = stats.sem(data)

# Calculate the 95% confidence interval
confidence_level = 0.95
degrees_of_freedom = len(data) - 1
critical_value = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)
margin_of_error = critical_value * std_error
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean:.4f}")
print(f"95% Confidence Interval: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")

# Interpret the confidence interval
print(f"We are 95% confident that the true population mean lies between {confidence_interval[0]:.4f} and {confidence_interval[1]:.4f}.")
```

```python
'''01. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and 
interpret the results'''
import numpy as np
from scipy.stats import norm

# Sample data
sample_mean = 15.2
sample_size = 50
population_mean = 14.5
population_std = 2.5

# Calculate Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

# Significance level (Î±)
alpha = 0.05

print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is statistically significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not statistically significantly different from the population mean.")
```

```python
'''04. Perform a two-tailed Z-test using Python and visualize the decision region on a plot'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
sample_mean = 10.5
population_mean = 10
population_std = 2
sample_size = 100
alpha = 0.05

# Calculate Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate critical Z-values
critical_z = norm.ppf(1 - alpha / 2)

# Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Plot standard normal distribution
x = np.linspace(-3, 3, 100)
y = norm.pdf(x, 0, 1)

plt.plot(x, y, color='blue')
plt.fill_between(x, y, where=(x < -critical_z) | (x > critical_z), color='red', alpha=0.3)
plt.axvline(x=-critical_z, color='red', linestyle='--')
plt.axvline(x=critical_z, color='red', linestyle='--')
plt.axvline(x=z_statistic, color='green', linestyle='-', label=f"Z-statistic = {z_statistic:.2f}")
plt.title('Standard Normal Distribution')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
'''05. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_errors(population_mean, sample_mean, population_std, sample_size, alpha):
    # Calculate Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate critical Z-values
    critical_z = norm.ppf(1 - alpha / 2)

    # Calculate Type 1 error (Î±)
    type_1_error = alpha
    # Calculate Type 2 error (Î²)
    beta = norm.cdf(critical_z - np.abs((sample_mean - population_mean) / (population_std / np.sqrt(sample_size)))) + (1 - norm.cdf(-critical_z - np.abs((sample_mean - population_mean) / (population_std / np.sqrt(sample_size)))))

    # Plot standard normal distribution
    x = np.linspace(-3, 3, 100)
    y = norm.pdf(x, 0, 1)
    plt.plot(x, y, color='blue')
    plt.fill_between(x, y, where=(x < -critical_z) | (x > critical_z), color='red', alpha=0.3, label='Type 1 Error')
    plt.axvline(x=-critical_z, color='red', linestyle='--')
    plt.axvline(x=critical_z, color='red', linestyle='--')
    plt.axvline(x=z_statistic, color='green', linestyle='-', label=f"Z-statistic = {z_statistic:.2f}")
    plt.title('Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()
    return type_1_error, beta

# Parameters
population_mean = 10
sample_mean = 11
population_std = 2
sample_size = 100
alpha = 0.05

type_1_error, type_2_error = calculate_errors(population_mean, sample_mean, population_std, sample_size, alpha)

print(f"Type 1 Error (a): {type_1_error:.4f}")
print(f"Type 2 Error (Î²): {type_2_error:.4f}")
```

```python
'''07. Perform a paired sample T-test using Python and visualize the comparison results'''
import numpy as np
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt

# Sample data (before and after treatment)
before = np.array([85, 88, 78, 92, 89, 76, 95, 89, 91, 82])
after = np.array([90, 92, 82, 95, 91, 80, 98, 92, 94, 85])

# Perform paired sample T-test
t_statistic, p_value = ttest_rel(before, after)

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Significance level (Î±)
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The treatment has a significant effect.")
else:
    print("Fail to reject the null hypothesis: The treatment has no significant effect.")
# Visualize the comparison result
plt.figure(figsize=(8, 6))
plt.plot(before, label='Before', marker='o')
plt.plot(after, label='After', marker='o')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.title('Comparison of Before and After Treatment')
plt.legend()
plt.show()

# Calculate and plot the differences
differences = after - before
plt.figure(figsize=(8, 6))
plt.plot(differences, label='Differences', marker='o')
plt.axhline(y=np.mean(differences), color='red', linestyle='--', label='Mean Difference')
plt.xlabel('Sample Index')
plt.ylabel('Difference')
plt.title('Differences Between Before and After Treatment')
plt.legend()
plt.show()
```

```python
'''08. Simulate data and perform both Z-test and T-test, then compare the results using Python'''
import numpy as np
from scipy.stats import norm, ttest_1samp

# Set the seed for reproducibility
np.random.seed(0)

# Simulate data
sample_size = 100
population_mean = 5
sample_data = np.random.normal(loc=population_mean, scale=1, size=sample_size)

# Calculate sample mean and standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data)

# Perform Z-test
z_statistic = (sample_mean - population_mean) / (1 / np.sqrt(sample_size))
p_value_z = 2 * (1 - norm.cdf(abs(z_statistic)))

# Perform T-test
t_statistic, p_value_t = ttest_1samp(sample_data, population_mean)

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Z-test: Z-statistic = {z_statistic:.4f}, p-value = {p_value_z:.4f}")
print(f"T-test: T-statistic = {t_statistic:.4f}, p-value = {p_value_t:.4f}")

# Compare the results
alpha = 0.05
if p_value_z < alpha:
    print("Z-test: Reject the null hypothesis")
else:
    print("Z-test: Fail to reject the null hypothesis")

if p_value_t < alpha:
    print("T-test: Reject the null hypothesis")
else:
    print("T-test: Fail to reject the null hypothesis")
```

```python
'''09. Write a Python function to calculate the confidence interval for a sample mean and explain its significance'''
import numpy as np
from scipy.stats import t

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    # Calculate sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    # Calculate standard error
    standard_error = sample_std / np.sqrt(sample_size)

    # Calculate critical t-value
    degrees_of_freedom = sample_size - 1
    critical_t = t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

    # Calculate margin of error
    margin_of_error = critical_t * standard_error

    # Calculate confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
sample_data = np.array([23, 21, 19, 24, 20, 22, 18, 25, 21, 19])
confidence_level = 0.95

lower_bound, upper_bound = calculate_confidence_interval(sample_data, confidence_level)

print(f"Sample Mean: {np.mean(sample_data):.4f}")
print(f"{confidence_level*100}% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
```

```python
'''10. Write a Python program to calculate the margin of error for a given confidence level using sample data'''
import numpy as np
from scipy.stats import t

# Sample data
sample_data = np.array([23, 21, 19, 24, 20, 22, 18, 25, 21, 19])

# Calculate sample mean and standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

# Sample size
sample_size = len(sample_data)

# Confidence level
confidence_level = 0.95

# Calculate standard error
standard_error = sample_std / np.sqrt(sample_size)

# Calculate critical t-value
degrees_of_freedom = sample_size - 1
critical_t = t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

# Calculate margin of error
margin_of_error = critical_t * standard_error

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Margin of Error at {confidence_level*100}% Confidence Level: {margin_of_error:.4f}")
print(f"Confidence Interval: ({sample_mean - margin_of_error:.4f}, {sample_mean + margin_of_error:.4f})")
```

```python
'''11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process'''
import numpy as np

# Prior probabilities
P_A = 0.01  # Probability of having a disease (A)
P_not_A = 1 - P_A  # Probability of not having a disease

# Conditional probabilities
P_B_given_A = 0.99  # Probability of testing positive given disease (B|A)
P_B_given_not_A = 0.05  # Probability of testing positive given no disease

# Calculate P(B) using the law of total probability
P_B = P_A * P_B_given_A + P_not_A * P_B_given_not_A

# Apply Bayes' Theorem to calculate P(A|B)
P_A_given_B = (P_B_given_A * P_A) / P_B

print(f"P(A|B) = {P_A_given_B:.4f}")
```

```python
'''12. Perform a Chi-square test for independence between two categorical variables in Python'''
import numpy as np
from scipy.stats import chi2_contingency

# Contingency table
observed = np.array([[80, 60], [70, 90]])

# Perform Chi-square test
chi2_statistic, p_value, dof, expected = chi2_contingency(observed)

print(f"Chi-square statistic: {chi2_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")

# Significance level (Î±)
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The variables are likely dependent.")
else:
    print("Fail to reject the null hypothesis: The variables are likely independent.")

print("\nObserved Frequencies:")
print(observed)
print("\nExpected Frequencies:")
print(expected)
```

```python
'''13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed 
data'''
import numpy as np

# Observed frequencies
observed = np.array([[80, 60], [70, 90]])

# Calculate total observations
total_observations = np.sum(observed)

# Calculate row and column totals
row_totals = np.sum(observed, axis=1)
column_totals = np.sum(observed, axis=0)

# Calculate expected frequencies
expected = np.outer(row_totals, column_totals) / total_observations

print("Observed Frequencies:")
print(observed)
print("\nExpected Frequencies:")
print(expected)
```

```python
'''14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution'''
import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed = [70, 80, 90, 60]

# Expected frequencies under a null hypothesis (e.g., equal frequencies)
expected = [75, 75, 75, 75]

# Perform chi-square goodness-of-fit test
chi2, p = chisquare(observed, expected)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("Reject the null hypothesis: Observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: Observed data fits the expected distribution.")
```
